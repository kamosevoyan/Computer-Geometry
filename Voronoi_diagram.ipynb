```python
import numpy
import matplotlib.pyplot as plt
from functions import voronoi_diagram
from scipy.spatial import Voronoi, voronoi_plot_2d
from scipy.spatial import Delaunay
```

```python
points = numpy.array([
[-4, 2],
[-3, 4],
[-1, 5],
[-1, 3],
[2, 2],
[2, 6],
[5, 5],
[8, -1]
])
```

# Delaunay triangulation

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.axis("equal")
fig = voronoi_plot_2d(Voronoi(points, ), ax=ax, color="cyan")
ax.scatter(points[:, 0], points[:, 1], color="red", s=100)
ax.triplot(points[:,0], points[:,1], Delaunay(points).simplices, "-.")
plt.show()
```
