```python
import numpy
import matplotlib.pyplot as plt
from functions import voronoi_diagram
```

```python
points = numpy.array([
[0, 1],
[1, 0],
[1, 2],
[-5, 5],
[-3, -3],
[0, -1],
[-6, 0],
[6, -5],
[6, 6]
])

points = points /  numpy.abs(points).max()
```

# Voronoi diagram

```python
result = voronoi_diagram(points)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(points[:, 0], points[:, 1], color="r", s=80)
ax.scatter(result[:, 0], result[: ,1], color="k", s=1)
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
plt.show()
```
