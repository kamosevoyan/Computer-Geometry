```python
import numpy
import matplotlib.pyplot as plt
from math import sqrt


from functions import polygonal_chain, ermit_spline, cubic_spline, norm
```

```python
points = numpy.array([
                        [-1,  1],
                        [ 1,  3],
                        [ 2, -1],
                        [-1,  2],
                        [ 0, -1],
                        [ 7, -2]
])
```

```python
tangents =  numpy.array([
                        [ 0,  1],
                        [ 1,  3],
                        [ 2,  1],
                        [ 1,  2],
                        [ 1, -1],
                        [ 1,  2]
])
```

```python
POINTS = ermit_spline(points, tangents)
```

```python
plt.figure(figsize=(13, 8))
plt.plot(POINTS[0], POINTS[1])
plt.scatter(points[:, 0], points[:, 1], c="r")
plt.title("Ermit spline")
plt.quiver(points[:, 0], points[:, 1], tangents[:, 0], tangents[:, 1], color=['r','b','g'], scale=20)
plt.show()
```

```python
POINTS = polygonal_chain(points)
```

```python
plt.figure(figsize=(13, 8))
plt.plot(POINTS[0], POINTS[1])
plt.scatter(points[:, 0], points[:, 1], c="r")
plt.title("Polygon chain")
plt.show()
```
