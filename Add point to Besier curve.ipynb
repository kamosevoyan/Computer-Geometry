```python
import numpy
import matplotlib.pyplot as pyplot
from functions import besier_curve, add_one_point
```

```python
points = numpy.array([
                        [0, 0],
                        [1, 0.3], 
                        [3, 0.2],
                        [4, .1]
                    
], dtype=float)

points_plus = add_one_point(points)

for it in range(0):
    points_plus = add_one_point(points_plus)   
```

```python
POINTS = besier_curve(points, density=1000)
POINTS_PLUS = besier_curve(points_plus, density=1000)
```

```python
pyplot.figure(figsize=(9,9))

pyplot.plot(POINTS[0, ...], POINTS[1, ...], "b")
pyplot.plot(POINTS_PLUS[0, ...], POINTS_PLUS[1, ...], "k")

pyplot.scatter(points[:, 0], points[:, 1], c="r")
pyplot.scatter(points_plus[:, 0], points_plus[:, 1], c="g")

pyplot.plot(points[:, 0], points[:, 1])
pyplot.plot(points_plus[:, 0], points_plus[:, 1])
pyplot.axis("auto")
pyplot.show()
```
