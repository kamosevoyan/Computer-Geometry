```python
import numpy
import matplotlib.pyplot as pyplot
from functions import polygonal_chain, cubic_spline, besier_curve
```

## Points

```python
points = numpy.array([
                        [6,   4],
                        [7,  10],
                        [9,  12],
                        [11,  2],
])
```

## Density

```python
density = 1000
```

## Getting (x,y) pairs from functions

```python
polygonal_chain_points = polygonal_chain(points, density=density)
cubic_spline_points = cubic_spline(points, density=density)
besier_curve_points = besier_curve(points, density=density)
```

## Plotting curves

```python
pyplot.figure(figsize=(12,9))

pyplot.plot(polygonal_chain_points[0, ...], polygonal_chain_points[1, ...])
pyplot.plot(cubic_spline_points[0, ...], cubic_spline_points[1, ...])
pyplot.plot(besier_curve_points[0, ...], besier_curve_points[1, ...])

pyplot.scatter(points[:, 0], points[:, 1], c="r")
pyplot.axis("tight")
pyplot.legend(["Polygonal chain", "Cubic spline", "Besier curve"])
pyplot.title("All three plots")
pyplot.show()
```
